```python
x = df.iloc[:,[3,4]].values              #it will select only the 3rd & 4th column
```

```python
print(x)
```

```python
sns.pairplot(df)       #it shows the scatterplot of the dataset
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(i) 
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
    
```

```python
print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=40)
```

```python
y_means  = kmeans.fit_predict(x)
```

```python
y_means
y_means.shape



```

```python
y_means
```

```python
plt.scatter(x[y_means==0,0],x[y_means == 0,1],c = 'red',label = "cluster1")
plt.scatter(x[y_means==1,0],x[y_means == 1,1],c = "black",label = "cluster2")
plt.scatter(x[y_means==2,0],x[y_means == 2,1],c = 'green',label = "cluster3")
plt.scatter(x[y_means==3,0],x[y_means == 3,1],c = 'yellow',label = "cluster4")
plt.scatter(x[y_means==4,0],x[y_means == 4,1],c = 'blue',label = "cluster5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c= "yellow")
```
