```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
movies = pd.read_csv("movie_metadata.csv")
```

```python
movies.tail()
```

```python
movies_1 = movies.dropna(axis=0,how = 'any')
movies_1.head(10)
```

```python
movies_1.drop(["color","director_name","duration","director_facebook_likes","actor_3_facebook_likes","actor_2_name","actor_1_facebook_likes","num_user_for_reviews","language","country","content_rating","budget","title_year","actor_2_facebook_likes","aspect_ratio","movie_facebook_likes","gross","actor_1_name","cast_total_facebook_likes","actor_3_name","actor_3_name","facenumber_in_poster","movie_imdb_link"],axis=1,inplace =True)
```

```python
movies_1.head()
```

```python
genres_sep = movies_1['genres'].str.get_dummies('|')
genres_sep.head()
```

```python
movie_df = pd.concat([movies_1, genres_sep], axis=1, sort=False)
movie_df.drop(["genres"],axis=1,inplace =True)
movie_df.head()
```

```python
movie_df["num_voted_users"].hist()  
```

```python
movie_df["num_critic_for_reviews"].hist(bins=20)      
```

```python
sns.jointplot(x="num_critic_for_reviews", y="num_voted_users" , data=movie_df) 
```
