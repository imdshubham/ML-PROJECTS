```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])   #if we provide a single list matplot lib assumes to be sequence of y-axis
plt.ylabel("numbers")
plt.xlabel("indices")
plt.title("myplot")
plt.show()
```

```python
plt.plot([4,2,6,4])   
plt.ylabel("numbers")
plt.xlabel("indices")
plt.title("myplot")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],"ro")
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^3')
plt.plot(t,t**2.5,'g',label='^3')
plt.grid()
plt.legend()
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth=5)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]

plt.plot(x1,y1,'r--')

plt.plot(x2,y2,'b--')
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)

plt.setp(lines[0],color='r',linewidth='2.0')

plt.setp(lines[1],'color','g','linewidth','2.0')

plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.,5.,0.1)
t2 = np.arange(0.,5.,0.2)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b--')
plt.show()

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
```
