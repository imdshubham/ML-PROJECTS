```python
def  addAge(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 38
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(addAge,axis=1)
```

```python
sns.heatmap(train.isnull(),cmap='viridis')
```

```python
train.drop('Cabin',axis=1,inplace=True)   #it will drop the cabin column from the data set
```

```python
sns.heatmap(train.isnull(),cmap='viridis')
```

```python
sex = pd.get_dummies(train['Sex'],drop_first = True)   #get_dummies will convert the categorical data into numerical form0
```

```python
sex.head()
```

```python
embarked = pd.get_dummies(train['Embarked'],drop_first = True)
```

```python
embarked.head()
```

```python
train1 = train.copy()    #copy method:- it will copy the dataset
```

```python
train1
```

```python
train = pd.concat([train,sex,embarked],axis=1)    #concat will concat the columns into the dataset
```

```python
train.head()
```

```python
train.drop(["PassengerId","Name","Embarked"],axis=1,inplace =True)   #using drop we can delete the unwanted columns
```

```python
train.head()
```

```python
train.drop(["Ticket","Sex"],axis=1,inplace =True)
```

```python
train.head()
```

```python
X = train.drop(["Survived"],axis=1)   #inplace = true is not use
```

```python
y =train['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression   #here we use logistic regression cause the outcome will be getting only in the  form of "yes" or "no".....i.e. if the no. of people will survive or die
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)   #on the above training model we will predict the testing model using the "predict"...
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))   #here we getting the 78% of the result which is reaaly good ....
```

```python
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions)) 
```
