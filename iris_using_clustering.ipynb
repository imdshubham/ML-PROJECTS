```python
df1=df.drop(["Species"],axis=1)
```

```python
sns.pairplot(df1) 
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(i) 
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
```

```python
print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=40)
```

```python
y_means  = kmeans.fit_predict(df1)
```

```python
y_means
y_means.shape



```

```python
y_means
```

```python
x=df1.iloc[:,[1,2,3,4]].values
x
```

```python
x.shape
```

```python
plt.scatter(x[y_means==0,0],x[y_means == 0,1],c = 'red',label = "cluster1")
plt.scatter(x[y_means==1,0],x[y_means == 1,1],c = "black",label = "cluster2")
plt.scatter(x[y_means==2,0],x[y_means == 2,1],c = 'green',label = "cluster3")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c= "yellow")
```
